# 练习

```c
#include<stdio.h>
int main(){
    
    // 冒泡排序
    int a[10]={6,3,4,7,8,9,0,4,5,3};
    int i,j,t;
    
    printf("排序前：");
    for (i = 0; i < 10; i++)
        printf("%d ", a[i]);
    printf("\n");
    for (i = 0; i < 9; i++){
        for (j = 0; j < 10-i-1; j++){
            if (a[j] > a[j+1]){
                t = a[j];
                a[j] = a[j+1];
                a[j+1]=t;
            }
        }
    }
    
    printf("排序后：");
    for (i = 0; i < 10; i++)
        printf("%d ", a[i]);
    
    return 0;
}


```
