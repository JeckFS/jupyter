# 练习

```c
#include<stdio.h>
int main(){
    
    // 冒泡排序
    int a[10]={6,3,4,7,8,9,0,4,5,3};
    int i,j,t;
    
    printf("排序前：");
    for (i = 0; i < 10; i++)
        printf("%d ", a[i]);
    printf("\n");
    for (i = 0; i < 9; i++){
        for (j = 0; j < 10-i-1; j++){
            if (a[j] > a[j+1]){
                t = a[j];
                a[j] = a[j+1];
                a[j+1]=t;
            }
        }
    }
    
    printf("排序后：");
    for (i = 0; i < 10; i++)
        printf("%d ", a[i]);
    
    return 0;
}


```

# 指针

* 何为指针？**指针即地址**。即存储单元的地址。
* 存储单元里面可以存储内容。
* 指针变量的值是地址，它是一个16进制的数字。

重点：<br><br>
    指针变量和指针：<br>
* 指针变量是存储另一个变量的指针的变量。它是一个变量，由于地址是一个16进制的整数，故而指针变量的存储内容是一个16进制的整数。
* 而指针就是一个地址值

指针符号：&(取地址)， *(取内容)

```c
#include <stdio.h>
void main(){
    int a = 1;
    int* p = &a;
    printf("指针变量存储的值：%p\n",p);
    printf("指针变量存储的内容：%d", *p);
    *p = 2;
    printf("%d\n");
}
```

传值调用与传址调用(引用调用)：


```c
// 传值调用：
#include <stdio.h>
void fun(int a, int b){
    int t;
    t = a;
    a = b;
    b = t;
    printf("fun...:\na=%d, b=%d\n",a,b);
}
int main(){
    int a = 1, b = 2;
    printf("main...:\na=%d, b=%d\n",a, b);
    fun(a,b);
    printf("main...:\na=%d, b=%d",a, b);
    return 0;
}
```

```c
// 传址调用
#include <stdio.h>
void fun(int* p, int* q){
    int t;
    t = *p;
    *p = *q;
    *q = t;
    printf("fun...:\na=%d, b=%d\n",*p, *q);
}
int main(){
    int a = 1, b = 2;
    printf("main...:\na=%d, b=%d\n",a, b);
    fun(&a, &b);
    printf("fun...:\na=%d, b=%d\n",a, b);
    return 0;
}
```

# 通过指针引用数组 (重点)

重要概念：<br>
数组指针：指向数组的指针<br>
数组元素指针：指向数组元素的指针

<font color='#f40c7a' size=3>数组名本身就是指针(地址)</font>

指针变量的运算：+ -<br>
* *(p++)  *(++p)
* ++(*p)
* *(p--) *(--p)  
<br>P237

通过指针引用数组元素：P233

数组名作为函数参数：P237

```c
// 利用指针实现数组元素的反转
# include <stdio.h>
void main(){
    int len = 10;
    int a[]={1,2,3,4,5,6,7,8,9,0};
    int* p = &a[0], *q = &a[len-1];
    
    int i;
    int t;
    
    while(q - p > 0){
        t = *q;
        *q = *p;
        *p = t;
        
        p++;
        q--;
    }
    
    for (i = 0; i < len; i++)
        printf("%d ",a[i]);
}
```
