# 线性表

线性表的存储结构：顺序表(逻辑连续，物理连续)，链表(逻辑连续，物理可连续可不连续)<br>
区别：<br>
    顺序表：空间不可扩充，访问速度为O(1)。等价于数组<br>
    链表：空间可扩充，访问速度为O(n)，需要遍历链表中的节点。适用于插入删除操作频繁的情况。<br>
