# 整理数据

```python
# 每种类别分别选取1000个样本的训练集，500个样本的验证集，
# 500个样本的测试集
# 并将这些数据集复制到对应的目录中
import os, shutil

original_dataset_dir='/home/yfs/桌面/dogs-vs-cats/train'
cur_dir=os.getcwd()
base_dir=os.path.join(cur_dir,"cats_and_dogs_small")
os.mkdir(base_dir)

# 创建训练集，验证集，测试集的目录
train_dir=os.path.join(base_dir,"train")
os.mkdir(train_dir)
validation_dir=os.path.join(base_dir,"validation")
os.mkdir(validation_dir)
test_dir=os.path.join(base_dir,"test")
os.mkdir(test_dir)

#创建猫的训练集目录
train_cats_dir=os.path.join(train_dir,"cats")
os.mkdir(train_cats_dir)

# 创建狗的训练集目录
train_dogs_dir=os.path.join(train_dir,"dogs")
os.mkdir(train_dogs_dir)

# 创建猫的验证集目录
validation_cats_dir=os.path.join(validation_dir,"cats")
os.mkdir(validation_cats_dir)

# 创建狗的验证集目录
validation_dogs_dir=os.path.join(validation_dir,"dogs")
os.mkdir(validation_dogs_dir)

# 创建猫的测试集目录
test_cats_dir=os.path.join(test_dir,"cats")
os.mkdir(test_cats_dir)

# 创建狗的测试集目录
test_dogs_dir=os.path.join(test_dir,"dogs")
os.mkdir(test_dogs_dir)


def copyfiles(kind,dst_dir,start,end):
    fnames=["{0}.{1}.jpg".format(kind,i) for i in range(start,end)]
    for fname in fnames:
        src=os.path.join(original_dataset_dir,fname)
        dst=os.path.join(dst_dir,fname)
        shutil.copyfile(src,dst)

        
# 复制猫的图像
copyfiles("cat",train_cats_dir,0,1000)
copyfiles("cat",validation_cats_dir,1000,1500)
copyfiles("cat",test_cats_dir,1500,2000)
# 复制狗的图像
copyfiles("dog",train_dogs_dir,0,1000)
copyfiles("dog",validation_dogs_dir,1000,1500)
copyfiles("dog",test_dogs_dir,1500,2000)
```

由于本问题是一个平衡分类问题(每个类别的可能性相同)，故可以使用精度作为衡量成功的指标。

# 构建网络

**网络中特征图的深度逐渐增大(从32增大到128)，而特征图的尺寸在逐渐减小(从150x150减小到7x7)。这几乎是所有卷积神经网络的模式。**

```python
from tensorflow.python.keras import layers
from tensorflow.python.keras import models

model=models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Flatten())
model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
# 编译
from tensorflow.python.keras import optimizers

model.compile(loss='binary_crossentropy',optimizer=optimizers.RMSprop(lr=1e-4),metrics=['acc'])
```

# 数据预处理与生成器拟合模型

预处理步骤：
1. 读取图像文件
2. 将JPEG文件解码为RGB像素网络
3. 将这些像素网络转化为浮点数张量
4. 将像素值(0~255)缩放到[0,1]区间

keras.preprocessing.image中的ImageDataGenerator类可以将硬盘上的图像文件自动转换为与处理好的张量批量。

```python
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator

# 将所有图像乘以1/255缩放
train_datagen=ImageDataGenerator(rescale=1./255)
test_datagen=ImageDataGenerator(rescale=1./255)

train_generator=train_datagen.flow_from_directory(
    train_dir, # 目标目录
    target_size=(150,150), # 将所有图像的大小调整为150x150
    batch_size=20, # 每个批量包含20个样本(一次抽取20个样本)
    class_mode='binary' # 因为使用了binary_crossentropy损失，所以需要用二进制标签
) 
validation_generator=test_datagen.flow_from_directory(
    validation_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='binary'
)
```

---

```python
history=model.fit_generator(
    train_generator, # 一个可以不停提供输入和目标组成的批量
    # 从生成器中抽取steps_per_epoch个批量(即运行了steps_per_epoch次梯度下降)
    # 每个批量包含20个样本，所以读完所有2000个样本需要100个批量
    steps_per_epoch=100, 
    epochs=30, 
    
    # 指定验证集(是一个生成器或者Numpy数组组成的元祖)
    validation_data=validation_generator,
    validation_steps=50 # 指定需要从验证生成器中抽取多少个批次用于评估     
)
```

```python
# 保存模型
model.save('cats_and_dogs_1.h5')
```

```python
# 绘制损失曲线与精度曲线
import matplotlib.pyplot as plt

acc=history.history['acc']
val_acc=history.history['val_acc']
loss=history.history['loss']
val_loss=history.history['val_loss']

epochs=range(1,len(acc)+1)

plt.plot(epochs,acc,'bo',label='Training acc')
plt.plot(epochs,val_acc,'b',label="Validation acc")
plt.title("Training and validation accuracy")
plt.legend()

plt.figure()

plt.plot(epochs,loss,'bo',label='Training loss')
plt.plot(epochs,val_loss,'b',label="Validation losss")
plt.title("Training and validation loss")
plt.legend()

plt.show()
```

# 使用数据增强降低过拟合

## 数据增强及简单用法

数据增强：是从现有的训练样本中生成更多的训练数据，方法是利用多种能够生成可信图像的随机变换来增加样本。

```python
datagen=ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
# rotation_range：角度值[0,180]，表示图像随机旋转的角度范围
# width_shift, height_shift：图像在水平或垂直方向上平移的范围(相对于总宽度和总高度的比例)
# shear_range：随机错切变换的角度
# zoom_range：图像随机缩放的范围
# horizontal_flip：随机将一半的图像水平翻转
# fill_mode：用于填充新创建像素的方法，这些新像素可能来自旋转或宽度/高度平移
```

**显示几个随机增强后的训练图像:**

```python
# 图像预处理工具的模块
from tensorflow.python.keras.preprocessing import image

fnames=[os.path.join(train_cats_dir,fname) for fname in os.listdir(train_cats_dir)]

# 选择一张图像进行增强
img_path=fnames[3]

# 读取图像并调整大小
img=image.load_img(img_path,target_size=(150,150))

# 将其转换为形状(150,150,3)的Numpy数组
x=image.img_to_array(img)

# 将其形状改变为(1,150,150,3)
x=x.reshape((1,)+x.shape)

i=0
for batch in datagen.flow(x,batch_size=1):
    plt.figure(i)
    imgplot=plt.imshow(image.array_to_img(batch[0]))
    i+=1
    if i%4==0:
        break
        
plt.show()
```

## 利用数据增强和Dropout降低过拟合

因为这些增强后的输入来自与少量的原始图像，它们具有高度相关性，不足以完全消除过拟合，为了保险再加一个Dropout层。

```python
model=models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer=optimizers.RMSprop(lr=1e-4),metrics=['acc'])
```

```python
# 利用数据增强生成器训练卷积神经网络
train_datagen=ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# 验证集不能数据增强
test_datagen=ImageDataGenerator(rescale=1./255)

train_generator=train_datagen.flow_from_directory(
    train_dir, # 目标目录
    target_size=(150,150), # 将所有图像的大小调整为150x150
    batch_size=32, # 批量大小为32(个样本)
    class_mode='binary' # 因为使用了binary_crossentropy损失，所以需要用二进制标签
)

validation_generator=test_datagen.flow_from_directory(
    validation_dir,
    target_size=(150,150),
    batch_size=32,
    class_mode='binary'
)

history=model.fit_generator(
    train_generator,
    steps_per_epoch=100,
    epochs=100,
    validation_data=validation_generator,
    validation_steps=50
)
```

```python
# 保存模型
model.save('cats_and_dogs_2.h5')
```

```python
history=history.history
history.keys()
```

```python
import matplotlib.pyplot as plt

train_acc=history['acc']
train_loss=history['loss']
val_acc=history['acc']
val_loss=history['val_loss']

epochs=range(1,len(train_acc)+1)

plt.plot(epochs, train_acc, 'bo', label='Training acc')
plt.plot(epochs,val_acc, 'b',label='Validation acc')
plt.legend()

plt.figure()
plt.plot(epochs,train_loss,'bo',label='Training loss')
plt.plot(epochs,val_loss,'b',label='Validation loss')
plt.legend()
plt.show()
```
